_Regression_

$$ Error = h(x) - y $$
$$ Square Error = (h(x) - y)^2  $$
$$ Mean Square Error =  ^1/_n * \sum_{}(h(x) - y)^2 $$


```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "data/HANBatang.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
import datetime

# 5월 1일부터 6월 11일까지 LG화학 주식의 저가, 고가에 대한 선형회귀를 해보자
df = pd.read_csv("data/marcap-2021.csv")
df = df[df["Date"] > "2021-05-01"]
df = df[df["Name"] == "LG화학"]
df = df[["High", "Low"]]
X = df["Low"]
y = df["High"]

# LinearRegression 모델을 생성하고, 그 안에 X, y 데이터를 fit 시킨다.
line_fitter = LinearRegression()
# fit() 메서드는 선형 회귀 모델에 필요한 두 가지 변수를 전달하는 것
line_fitter.fit(X.values.reshape(-1, 1), y)
'''
여기서 주의해야 할 점은 X데이터를 넣을 때
values.reshape(-1,1)를 해줬다는 것
왜냐하면 X는 2차원 array 형태여야 하기 때문
이런 식으로 [[x1], [x2], [x3], ... , [xn]]
(이렇게 넣는 이유는 X 변수가 하나가 아니라 여러개일 때 다중회귀분석을 실시하기 위함
'''

plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.grid()
plt.xlabel('저가')
plt.ylabel('고가')
plt.show()
```
